```python
import os
import sys
from datetime import datetime
import pytz
import h5py
import pandas as pd
import csv
import numpy as np
from scipy.signal import medfilt
import matplotlib.pyplot as plt

def openFile():
    file = 'CSVFiles/csv_all_files.csv'
    file = open(file, "r")
    df = pd.read_csv(file)
    return df

def search_by_complex_queries(df, dataset):
    dataset = dataset.split(' ')
    try:
        query = dataset[0]
        op = dataset[1]
        val = dataset[2]
        datasets = search_by_dataset_name(df, query)
        final_ans = []
        for row in datasets.iterrows():
            if(op == ">") :
                if(float(row[1]['SingletonValue']) > float(val)):
                    final_ans.append(row)
            elif (op == "<") :
                if(float(row[1]['SingletonValue']) < float(val)):
                    final_ans.append(row)
    except :
        print ("Not a Valid Query")
    return final_ans


def search_by_dataset_name(df, dataset):
    df = df[df['DatasetName'].str.contains(str(dataset))==True]
    if(len(df) == 1):
        filename = df['DatasetName'].values[0]
        parentGroup = df['GroupName'].values[0]
        H5File = df['HDF5FileName'].values[0]
        file = h5py.File("H5Files"+"/"+str(H5File), 'r')
        parentGroup = file[parentGroup]
        return get_dataset(file, parentGroup)
    else:
        return df
    
def get_dataset(h5File, parentGroup):
    dataset = {'DatasetName' : h5File.name}
    for file in parentGroup.keys() :
        values = parentGroup[file].name.split("/")
        val = list(parentGroup[file])
        if len(val) == 1 :
            dataset.update({values[len(values)-1] : val[0]})
        else :
            dataset.update({parentGroup.name+"/"+str(values[len(values)-1]) : list(val)})
    return dataset

df = openFile()
val = search_by_dataset_name(df, "streakImage")
# val = search_by_dataset_name(df, "/AwakeEventData/XMPP-STREAK/StreakImage/streakImageData")
val = search_by_dataset_name(df, "/AwakeEventData/TT41.BCTF.412340/Acquisition/totalCurrentPreferred")
#print (val)
val = search_by_complex_queries(df, "/AwakeEventData/TT41.BCTF.412340/Acquisition/totalCurrentPreferred > 0.06")
print (val)
```
