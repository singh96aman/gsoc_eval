```python
import os
import h5py
from datetime import datetime
import pytz
import csv

def show_dataset(dataset):
    print ("Dataset successfully searched and created. Here are all the keys :")
    for row in dataset.items() :
        print (row)
    
def get_dataset(h5File, parentGroup):
    dataset = {'DatasetName' : h5File.name}
    for file in parentGroup.keys() :
        values = parentGroup[file].name.split("/")
        val = list(parentGroup[file])
        if len(val) == 1 :
            dataset.update({values[len(values)-1] : val[0]})
        else :
            dataset.update({parentGroup.name+"/"+str(values[len(values)-1]) : list(val)})
    return dataset
    

def search_dataset(dir_name, dataset_name):
    with open('CSVFiles/csv_final.csv', mode='r') as csv_file :
        reader = csv.reader(csv_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
        isHeader = True
        dataset = []
        for row in reader :
            if isHeader :
                isHeader = False
            else :
                if row[0] == dataset_name :
                    dataset = row
                    print (dataset)
                    break
        h5file = h5py.File(dir_name+"/"+str(dataset[2]), 'r')
        groups = dataset_name.split("/")
        groups = groups[1:]
        for group in groups :
            parentGroup = h5file
            h5file = h5file[group]
        return (h5file, parentGroup)

def create_CSV():
    dir_name = "H5Files"
    file_list = get_all_filenames(dir_name)
    for file in file_list :
        if file[-3:] == ".h5" :
            utc_dt, cern_dt, fileptr1, fileptr2 = convert_unix_time_to_datetype(file)
            h5file = h5py.File(dir_name+"/"+str(file), 'r')
            CSVMapping = []
            CSVMapping = get_csv_mapping(CSVMapping, h5file, file)
            write_into_csv(CSVMapping)
            print ("WROTE IN CSV")
                
def convert_unix_time_to_datetype(filename) :
    filename_array = filename.split("_")
    if len(str(filename_array[0])) == 19 :
        timestamp = float(filename_array[0])
        utc_dt = datetime.utcfromtimestamp(timestamp // 1e9)
        cern_tz = pytz.timezone('Europe/Zurich')
        cern_dt = utc_dt.replace(tzinfo=pytz.utc).astimezone(cern_tz)
        cern_dt = cern_tz.normalize(cern_dt)
        return (utc_dt, cern_dt, filename_array[1], filename_array[2])

def write_into_csv(CSVMapping) :
    with open('CSVFiles/csv_final.csv', mode='w') as csv_file :
        fieldnames = ['DatasetName', 'GroupName', 'FileName','Size', 'Shape', 'Datatype']
        print (fieldnames)
        writer = csv.writer(csv_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
        writer.writerow(fieldnames)
        for row in CSVMapping :
            if len(row) != 0 :
                writer.writerow(row)

def get_csv_mapping(CSVMapping, file, filename) :
    if(isinstance(file, h5py.Group)) :
        for sub in file.keys() :
            if(isinstance(file[sub], h5py.Dataset)) :
                if (file[sub].size == 0):
                    CSVMapping.append([file[sub].name, file.name, filename, file[sub].size, file[sub].shape, "NA"])
                else :
                    try:
                        CSVMapping.append([file[sub].name, file.name, filename, file[sub].size, file[sub].shape, str(file[sub].dtype)])
                    except :
                        CSVMapping.append([file[sub].name, file.name, file[sub].size, file[sub].shape, "NA"])                        
            elif (isinstance(file[sub], h5py.Group)) :
                get_csv_mapping(CSVMapping, file[sub], filename)
    return CSVMapping

def get_all_filenames(filepath) :
    return os.listdir(filepath)

create_CSV()
h5File, parentGroup = search_dataset("H5Files", "/AwakeEventData/XMPP-STREAK/StreakImage/streakImageData")
dataset=get_dataset(h5File, parentGroup)
show_dataset(dataset)
```
